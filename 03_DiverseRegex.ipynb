```python
data = 'From Stephen.Marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
atpos = data.find('@')
print('atpos:', atpos)
sppos = data.find(' ', atpos) # find spoce after atpos by giving argument atpos
print('sppos:', sppos)
host = data[atpos+1 : sppos]
print(host)
```

#### double split pattern

```python
data = 'From Stephen.Marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
words = data.split()
email = words[1]
pieces = email.split('@')
print(pieces[1])
```

#### double split pattern with re

```python
import re
lin = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('@([^ ]*)', lin)
print(y)
```

[^ ] 는 빈칸이 아닌 character와 match가 되고 *는 몇개든 제한 없다는 뜻이다.

#### 정교화

```python
import re
lin = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('^From .*@([^ ]*)', lin)
print(y)
```

#### Spam detection (실습)

```python
import re
hand = open('mbox-short.txt')
numlist = []
for line in hand:
    line = line.rstrip()
    stuff = re.findall('^X-DSPAM-Confidence: ([0-9.]+)', line)
    if len(stuff) != 1 : continue
    num = float(stuff[0])
    numlist.append(num)
print('Maximum:', max(numlist))
```
