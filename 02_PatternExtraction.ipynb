#### re.search()는 해당 문자열이 대상 정규식을 만족시키는지를 True/False로 리턴해주는 bool 값이다. 만약 매칭된 문자열을 추출하고 싶으면 re.findall()을 사용하면 된다.

```python
import re
x = 'My 2 favorite numbers are 19 and 42'
y = re.findall('[0-9]+', x)
print(y)
```

정규식에서 [0-9]는 문자 클래스를 의미한다. 0부터 9 까지 딱 한글자인 숫자 하나를 의미한다. +는 바로 앞 패턴이 1번 이상 반복된 것을 의미하므로, [0-9]+는 '연속된 숫자 덩어리 하나'를 의미한다.

```python
import re
x = 'My 2 favorite numbers are 19 and 42'
y = re.findall('[AEIOU]+', x)
print(y)
```

AEIOU에 해당하는 한 글자 이상의 문자가 없으므로 아무것도 반환하지 않는다.

#### Greedy Matching

```python
import re
x = 'From: Using the : character'
y = re.findall('^F.+:', x)
print(y)
```

'From:' 이 아닌 이유는 characters (*과 +)는 가장 길게 매칭되는 경우를 검색하기 때문이다.

#### Non-Greedy Matching

```python
import re
x = 'From: Using the : character'
y = re.findall('^F.+?:', x)
print(y)
```

?를 붙여주면 욕심을 버리고 가장 짧은 경우를 검색한다.

#### 섬세한 문자열 추출

```python
x = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('\S+@\S+', x)
print(y)
```

```python
x = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
y = re.findall('^From (\S+@\S+)', x)
print(y)
```

괄호는 매칭에 비포함되고, 추출될 문자열의 시작지점과 끝지점을 지정해주는 역할을 한다.
