#### Without re

```python
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if line.find('From:') >= 0:
        print(line)
```

```python
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if line.startswith('From:'):
        print(line)
```

#### With re

```python
import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip() # rstrip()은 문자열 오른쪽 끝의 공백 문자들을 제거해준다.
    if re.search('From:',line):
        print(line)
```

```python
import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:', line):
        print(line)
```
